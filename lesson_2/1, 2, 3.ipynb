```python
import numpy as np
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))

def binomial(n, k, p):
    return combinations(n, k) * (p**k) * ((1-p)**(n-k))
```

```python
#случайное распределение
list_d = []
for k in range(6):
    x = binomial(5, k, 0.25)
    list_d.append(x)
print(list_d)
```

```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
%config InlineBackend.figure_formats = ['svg']
plt.plot(list_d)
```

```python
#мат ожидание
x_probabilities = np.array(list_d)
x_values = np.arange(0, 6)
m = (x_values * x_probabilities).sum()

print(m) #ожидаем, что студент угадает 1 вопрос из 5 с вероятность 0.4
```

```python
#дисперсия случайной величины
y_values = x_values - m

z_values = y_values ** 2

d = (z_values * x_probabilities).sum()

print(d)
```

```python
def poisson_proba(k: int, lambda_: float) -> float:    
    return (lambda_ ** k) * (np.exp(-lambda_)) / np.math.factorial(k)
```

```python
lambda_ = 10
for j in range(20):
    x = sum(poisson_proba(k=i, lambda_=lambda_) for i in range(j))
    if x >= 0.95:
        print(f'с вероятнотсью: {x}')
        print(f'пользователь получит не более {j} писем в день')
        break
```
